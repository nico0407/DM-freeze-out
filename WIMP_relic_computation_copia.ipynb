$$\Omega_{\chi}h^2= \frac{\rho_{\chi}^0h^2}{\rho_c^0}= \frac{m_{\chi}s_0 Y_0 h^2}{\rho_c^0} \ \ \ \ \ \ \ \ \ \  s_0= \frac{2 \pi^2}{45}g_{*s}(T_{\gamma}) T_{\gamma}^3$$

$$\frac{s_0 h^2}{\rho_c^0}= \frac{3120 \times 0.71^2}{1.05375 \times 10^{-5}} \approx 1.5 \times 10^{8}$$

$$\Omega_{\chi}h^2= 1.5 \times 10^{8} m_\chi Y_0 (c^2/ \text{GeV})= 1.5 \times 10^{10} Y_0 \Big( \frac{m_\chi}{100 \text{GeV}/c^2} \Big)\rightarrow = 0.118$$

```python
def friz(xilof):
    yota = np.empty_like(xilof)
    mask1 = (0 <= xilof) & (xilof <= 1)
    mask2 = (1 < xilof) & (xilof <= 2)
    mask3 = np.logical_not((0 <= xilof) & (xilof <= 2)) #or ~((0 <= x) & (x <= 2))
    yota[mask1] = xilof[mask1]
    yota[mask2] = 2 - xilof[mask2]
    yota[mask3] = 0
    return yota

xlist = np.linspace(-1, 3, num = 1000)
ylist = friz(xlist)

plt.plot(xlist, ylist)
plt.show()
```
