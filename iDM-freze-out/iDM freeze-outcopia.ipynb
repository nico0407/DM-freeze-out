# Define the equilibrium distribution for $Y_1$ and $Y_2$

```python
def Yequilibrium(x):
    #T=m/x
    yota = np.empty_like(x)
    
    point1 = (m/x <= 0.01668479)
    point2 = (0.01668479 < m/x) & (m/x <= 0.11748257)
    point3 = (0.11748257 < m/x) & (m/x <= 0.22748515)
    point4 = (0.22748515 < m/x) & (m/x <= 8.6167)
    point6 = (8.6167 < m/x) & (m/x <= 162.48)
    point5 = (162.48 < m/x)
    
    yota[point1] = 0.145*x[point1]**(3/2)*np.exp(-x[point1])*2/g_star1(m/x[point1])
    yota[point2] = 0.145*x[point2]**(3/2)*np.exp(-x[point2])*2/g_star2(m/x[point2])
    yota[point3] = 0.145*x[point3]**(3/2)*np.exp(-x[point3])*2/g_star3(m/x[point3])
    yota[point4] = 0.145*x[point4]**(3/2)*np.exp(-x[point4])*2/g_star4(m/x[point4])
    yota[point6] = 0.145*x[point6]**(3/2)*np.exp(-x[point6])*2/g_star6(m/x[point6])
    yota[point5] = 0.145*x[point5]**(3/2)*np.exp(-x[point5])*2/g_star5(m/x[point5])
    return yota

# THIS ONE CAN BE PLOTTED BUTH NOE USED FOR INTEGRATION
```

# Implement the iDM freeze-out

$$ \frac{dY_{1,2}}{dx} = - \frac{\lambda_A}{x^2} \Big[ Y_1 Y_2 - Y_1^{(0)}Y_2^{(0)} \Big] \pm x \lambda_D \Big( Y_2- \frac{Y_2^{(0)}}{Y_1^{(0)}}Y_1 \Big) \pm \frac{\lambda_S}{x^2}Y_f^{(0)} \Big( Y_2- \frac{Y_2^{(0)}}{Y_1^{(0)}}Y_1 \Big) $$

```python
def odesiDM(x,t):
    t2=t*((m1+delta)/m1)
    
    #assigng each ODE to a vector element
    #Y1=x[0]
    #Y2=x[1]
    
    #nell'equazione di sopra integro dY in dx, dove probebilmente nel paper di gordon  stato scelto
    #essere x2, quindi integro in x2 che sarebbe t2 nell'equazione di python in questo caso.
    
    #define each ODE
    dY1dt= -k*m1*(Y1*Y2-Yequil(t)*Yequil(t2))/t2**2 #+ ks*m1*Yeq(t)*(Y2-Y1)/t**2 + t*kd*m1*(Y2-Y1)
    dY2dt= -k*(m1+delta)*(Y1*Y2-Yequil(t)*Yequil(t2))/t2**2 #- ks*m2*Yeq(t)*(Y2-Y1)/t**2 - t*kd*m2*(Y2-Y1)
    
    return [dY1dt, dY2dt]
```
