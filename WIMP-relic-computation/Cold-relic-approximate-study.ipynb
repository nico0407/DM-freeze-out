```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy import optimize
import pandas as pd

from scipy import integrate
from scipy import special
```

```python
def function(x):
    return np.sqrt(x)*np.exp(-x)
```

```python
x=np.linspace(0.2,60,1000)
```

```python
plt.plot(x,function(x),"k-")
plt.axhline(1E-14,color="red",linestyle='dashed')


plt.yscale('log')
plt.title(r'$\sqrt{x} \cdot e^{-x}$', size= 15)
plt.xlabel('$x$',size=15)
#plt.ylabel(r'$\sqrt{x} \cdot e^{-x}$',size=15)
plt.xlim(0.2,50)
#plt.ylim(1E-20,0.01)
plt.grid()
```
