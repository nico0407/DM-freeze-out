```python
from scipy import integrate
```

```python
# parameters
m = 100                                #GeV DM Mass
Mp = 1.22e19                            #GeV Planck Mass
```

$$\frac{dY}{dx}=- \Bigl(\frac{45}{\pi M_P^2}\Bigr)^{-1/2} \frac{g_*^{1/2}m}{x^2} \langle\sigma v\rangle(Y^2-Y_{eq}^2)$$

```python
def Yeq(x):
    return 0.145*x**(3/2)*np.exp(-x)
```

```python
global k
k=1
fk=lambda sigmav: np.sqrt(np.pi*(Mp**2)/45)*m*sigmav  #starting to build the dY/dx

def BEx(x,m=100):
    T=m/x
    if T< 0.00103107:
        return k*np.sqrt(g_star1(T))/x**2   #missing piece of the formula.
    elif 0.01874165<T<0.11748257:
        return k*np.sqrt(g_star2(T))/x**2
    elif 0.12337614<T<0.22748515:
        return k*np.sqrt(g_star3(T))/x**2
    elif 0.23169929<T<105.25139665:
        return k*np.sqrt(g_star4(T))/x**2
    else:
        return k*np.sqrt(g_star5(T))/x**2
```

```python
def BE(Y,x):
    return -(Y**2-Yeq(x)**2)*BEx(x)   #putting together the things to have the right formula

#THESE FUNCTIONS HAVE BEEN CREATED TO ACHIVE THE COMPLETE FORMULA FOR THE dY/dx
```

```python
xx=np.logspace(np.log10(1),np.log10(1000),1000)
plt.loglog(xx,Yeq(xx),'m',label=r'$Y_{\rm eq}$')




k=fk(sigmav=1E-18)  
x=np.logspace(np.log10(1),np.log10(1000),10000)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'r--',label=r'$\langle \sigma v \rangle = 10^{-18}\ {\rm GeV}^{-2}$')

k=fk(sigmav=1E-17)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

k=fk(sigmav=1E-16)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

k=fk(sigmav=1E-15)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

k=fk(sigmav=1E-14)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

k=fk(sigmav=1E-13)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

k=fk(sigmav=1E-12)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

k=fk(sigmav=1E-11)
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

k=fk(sigmav=1E-10) 
Y=integrate.odeint(BE,Yeq(x[0]),x)
plt.loglog(x,Y,'k:')

#WHAT HAPPENS WHEN THE AVERAGE CROSS SECTION*VELOCITY DROPS BELOW 1E-10?!?!?!?!?!?!?




plt.hlines(1E-11,1,1000,linestyles='-',color='g',lw=2,label='$Y_0$')    #THAT'S THE HORIZONTAL LINE IN PURPLE. 
                                                                #REFERS TO THE RIGHT EXPERIMENTAL ABOUNDANCE  


plt.ylim(1E-14,1)
plt.xlim(1,1000)
plt.xlabel('$x = m/T$', size= 15)
plt.ylabel('$Y$', size= 15)
plt.title('$m = 100$ GeV', size= 15)
plt.legend()
plt.grid()

# x grows in time means that the temperature decrease
```
